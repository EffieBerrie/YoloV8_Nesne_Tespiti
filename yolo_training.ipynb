Bu rapor, "Kupa" ve "Kalem" nesnelerinin tespiti için gerçekleştirilen model eğitiminin sonuçlarını içerir.

```python
import cv2
import matplotlib.pyplot as plt
import os

# Eğitim sonuçları grafiğini (results.png) ekrana getirme
resim_yolu = r"C:\Users\Owner\runs\detect\train4\results.png"

if os.path.exists(resim_yolu):
    img = cv2.imread(resim_yolu)
    plt.figure(figsize=(18, 12))
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.title("Eğitim Kayıp (Loss) ve Başarı (mAP) Grafikleri")
    plt.axis('off')
    plt.show()
else:
    print("Grafik dosyası bulunamadı.")
```

```python
# Confusion Matrix (Modelin hatalarını gösteren tablo)
matrix_yolu = r"C:\Users\Owner\runs\detect\train4\confusion_matrix.png"

if os.path.exists(matrix_yolu):
    img = cv2.imread(matrix_yolu)
    plt.figure(figsize=(10, 8))
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.title("Confusion Matrix")
    plt.axis('off')
    plt.show()
else:
    print("Confusion Matrix görseli bulunamadı.")
```
